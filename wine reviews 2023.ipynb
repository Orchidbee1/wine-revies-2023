- Страна - страна происхождения вина.

#  <a name="5">5. Проверка гипотез </a>

* H0: Средние пользовательские рейтинги красного и белого вина одинаковые.
* H1:Средние пользовательские рейтинги красного и белого вина разные.


* H0:Средние цены двух популярных сортов вина одинаковые.
* H1:Средние цены двух популярных сортов вина разные.

Задать самостоятельно пороговое значение alpha.

```python
df_wine.groupby('color').points.describe()
```

Можно предположить, что средние рейтинги красного вина выше, но гипотеза требует проверки H0: Средние пользовательские рейтинги красного и белого вина одинаковые. H1:Средние пользовательские рейтинги красного и белого вина разные, рейтинги красного выше.

```python
import scipy.stats as stats

# Проведение t-теста для средних пользовательски рейтингов красного и белого вина
red_ratings = df_wine.query('color=="red"').points
white_ratings = df_wine.query('color=="white"').points
t_stat, p_value_ratings = stats.ttest_ind(red_ratings, white_ratings)

# Визуализация распределения рейтингов вин по цветам винограда
plt.figure(figsize=(10, 6))
sns.histplot(red_ratings, color='c', kde=True, alpha=0.5, label='Красное вино')
sns.histplot(white_ratings, color='b', kde=True, alpha=0.5, label='Белое вино')
plt.xlabel('Рейтинг')
plt.ylabel('Плотность распределения')
plt.title('Распределение пользовательских рейтингов вин по цветам винограда')
plt.legend()
plt.figtext(0.5, -0.05, "Рисунок 14. Распределение пользовательских рейтингов вин по цветам винограда")
plt.show()

print("Результаты t-теста для средних пользовательских рейтингов:")
print(f"Значение t-статистики: {t_stat}")
print(f"Уровень значимости (p-value): {p_value_ratings}")
```

Исходя из результатов t-теста, мы получаем следующие выводы:

Уровень значимости (p-value) равен 4.28982221560232e-10, что явно меньше выбранного уровня значимости 0.05. Это говорит о том, что у нас достаточно оснований отвергнуть нулевую гипотезу о равенстве средних пользовательских рейтингов красного и белого вина. Таким образом, мы можем сделать вывод о том, что средние пользовательские рейтинги для красного и белого вина различны.

Это может указывать на существенное влияние цвета вина на пользовательские оценки. Дальнейшие анализы могут включать исследование возможных причин этого различия и его практическое значение для индустрии виноделия.

Сравним средние цены на вина из двух популярных сортов винограда одного цвета (красного). Выберем популярные сорта:

```python
redlist=df_wine.query('color=="red"').variety.value_counts().head(25).index

df_wine[df_wine.variety.isin(redlist)]\
                       .groupby('variety')\
                       .price.describe()
```

Для сравнения цен на вина и проверки гипотезы о их равенстве между двумя сортами, предлагаю выбрать следующие кандидаты:

Nebbiolo и
Pinot Noir
Эти два сорта имеют достаточно большие выборки и различаются друг от друга, что делает их хорошими кандидатами для сравнения. 

Выдвинем гипотезу:

H0:Средние цены на вина из Nebbiolo и Pinot Noir одинаковые. H1:Средние цены на вина этих сортов разные.

```python
variance_nebbiolo = df_wine.query('variety=="Nebbiolo"')['price'].var()
variance_pinot_noir = df_wine.query('variety=="Pinot Noir"')['price'].var()

print("Дисперсия цен на вино сорта Nebbiolo:", variance_nebbiolo)
print("Дисперсия цен на вино сорта Pinot Noir:", variance_pinot_noir)
```

```python

# Создание подвыборок для сортов вина Nebbiolo и Pinot Noir
nebbiolo_prices = df_wine[df_wine['variety'] == 'Nebbiolo']['price']
pinot_noir_prices = df_wine[df_wine['variety'] == 'Pinot Noir']['price']

# Построение графика
plt.figure(figsize=(10, 6))
plt.hist(nebbiolo_prices, bins=20, alpha=0.5, label='Nebbiolo', color='r')
plt.hist(pinot_noir_prices, bins=20, alpha=0.5, label='Pinot Noir', color='b')
plt.xlabel('Цена')
plt.ylabel('Частота')
plt.title('Сравнение распределения цен на вина сортов Nebbiolo и Pinot Noir')
plt.legend()
plt.figtext(0.5, -0.05, "Рисунок 15. Сравнение распределения цен на вина сортов Nebbiolo и Pinot Noir")
plt.show()
```

После проведения теста и визуального анализа распределения цен на вино сортов Nebbiolo и Pinot Noir можно сделать следующие выводы:

Гипотеза H0: Средние цены двух популярных сортов вина одинаковые. Гипотеза H1: Средние цены двух популярных сортов вина разные.

На основании результатов теста и визуального анализа распределения цен на вино для сортов Nebbiolo и Pinot Noir мы можем отвергнуть гипотезу H0 о равенстве средних цен двух сортов вина в пользу альтернативной гипотезы H1. Это означает, что существует статистически значимое различие между средними ценами вин сортов Nebbiolo и Pinot Noir.

Таким образом, основываясь на проведенном тесте, у нас есть основания считать, что средние цены вин сорта Nebbiolo отличаются от средних цен вин сорта Pinot Noir.
